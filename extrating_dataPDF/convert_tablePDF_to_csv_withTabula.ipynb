   # Usando tabula para converter tabelas pdf to dataframe pandas

```python
import tabula as tb
import pandas as pd
```

```python
file_input = "pdfs/result_enare.pdf"
```

```python
# Configure the area of the present table
fc = 28.35 # fator de conversão A4 
top = 0.1*fc 
left = 0.2*fc
bottom = 29*fc
width = 15*fc

# Read the pdf and make the area for extract data and export to a csv file
tb.convert_into(file_input, 'result.csv', output_format="csv",
                   lattice = True,
                   pages=['260-298'],
                   area= (top, left, bottom, width))
```

```python
# Name for columns
col=['Nome', 'ID', 'Cirurgia Geral',
         'Clinica Medica', 'Pediatria',
         'Gineco.Obs', 'Med.Prev.Soci',
         'Nota_Obj']
```

### CheckPoint - Carga dos dados

```python
data = pd.read_csv('result.csv')
```

```python
# Deletando a coluna Unnamed: 8 
data.drop('Unnamed: 8', axis = 1, inplace=True)
data
```

```python
# Renomeando os nomes das variáveis na coluna 
data.columns = col
```

### Tratamento do dataframe

```python
# Excluindo algumas linhas que não pertencem ao dataframdata
data.drop(labels=range(0,30), axis=0, inplace=True)
data.head()
```

```python
# Verificando o final do dataframe 
data.tail(50)
```

```python
# Exluindo linhas do final do dataframe
data.drop(labels=range(2565, 2602), axis=0, inplace= True)
```

```python
data.tail()
```

```python
data.info()
```

```python
# Verificando a presença de na no dataframe
data.isnull().sum()
```

```python
# Localizando onde está a observação com valor na
data[data['ID'].isnull()]
```

```python
# Removendo o valor na
data.dropna(inplace=True)
data.isna().sum()
```
